```python
imp_df = data.drop(columns = ['index','sorted_index','feature','type'])
imp_df
```

## KMeans

```python
k_values = np.array([3])
for k in k_values:
    km = sk.cluster.KMeans(n_clusters = k, init='random',n_init=1,max_iter=300,tol=1e-04,random_state=0)
    y_km = km.fit_predict(imp_df)

    centers = km.cluster_centers_
    labels = km.labels_
    inertia = km.inertia_
    print('Number of Clusters = ',k)
    print('--------------------------------------------------------------')
    print('Cluster Centers: ',centers)
    print('Labels: ',labels)
    print('Unique Labels: ',np.unique(labels))
    
    fig = plt.figure(figsize=(26,6))
    ax = fig.add_subplot(131, projection='3d')
    scatter = ax.scatter(imp_df['mRNA_fscore'], imp_df['siRNA_fscore'], imp_df['RNA_fscore'], c=y_km, s=15)
    ax.set_xlabel('mRNA fscore')
    ax.set_ylabel('siRNA fscore')
    ax.set_zlabel('Mixed fscore')
    #plt.legend(*scatter.legend_elements())
    #ax.legend(labels = [i for i in np.arange(k)])
    #ax.legend()
    plt.show()
    #plt.scatter(X[:, 1], X[:, 0], c=y_km, s=50, cmap='viridis')
    #plt.scatter(centers[:, 1], centers[:, 0], c='black', s=50, alpha=1);
    print('--------------------------------------------------------------')
```

## KMeans Elbow Plot

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = sk.cluster.KMeans(n_clusters=k)
    km = km.fit(imp_df)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
data_clustered = data.copy()
data_clustered = data_clustered.drop(columns = ['index','sorted_index'])
data_clustered['clusters'] = y_km

#map for custom sorting
sort_map = {2: 0, 1: 2, 0: 1, 3 : 3}
data_clustered = data_clustered.sort_values(by=['clusters'], key=lambda x: x.map(sort_map)).reset_index(drop = True)
#data_clustered.to_excel(os.path.join(featImpDir, 'Featimport_kmeans.xlsx'))
display(data_clustered)
```

## Principal Component Analysis

```python
imp_df = imp_df.iloc[:4,:]
```

```python
imp_df
```

```python
x = StandardScaler().fit_transform(imp_df)
pca = PCA(n_components=2)
components = pca.fit_transform(x)
pca_df = pd.DataFrame(data = components, columns = ['PC1', 'PC2'])
#pca_df.to_excel(os.path.join(featImpDir, 'Featimport_PCA.xlsx'))
pca_df
```

```python
fig, ax = plt.subplots()
ax.scatter(pca_df['PC1'], pca_df['PC2'], cmap='blues')
plt.ylabel('Principal Component 2')
plt.xlabel('Principal Component 1')
plt.show()
```

```python
print('explained variance: ',pca.explained_variance_ratio_)
print('singular values: ',pca.singular_values_)

# identity matrix
identity = np.identity(imp_df.shape[1])  
coef = pca.transform(identity)
print(' ')
print('coefficients: ')
print(coef)
```

```python
corr_df = data.drop(columns = ['index','sorted_index','feature'])
corr_df
```

```python
types = np.unique(corr_df['type'])
print(types)
```

```python
X = np.arange(3)
#fig = plt.figure()
#ax = fig.add_axes([0,0,1,1])
fig, axs = plt.subplots(6, 1, figsize = (10,16))
fig.tight_layout()
for i, feat_type in enumerate(types):
    #print('----------------------------------'+feat_type+' Features----------------------------------')
    mRNA = corr_df.loc[corr_df['type'] == feat_type,'mRNA_fscore']
    siRNA = corr_df.loc[corr_df['type'] == feat_type,'siRNA_fscore']
    mixed = corr_df.loc[corr_df['type'] == feat_type,'RNA_fscore']

    #Correlation
    pear_corr_mRNA_siRNA, _ = pearsonr(mRNA, siRNA)
    #print('mRNA/siRNA Pearsons correlation: %.3f' % pear_corr_mRNA_siRNA)
    pear_corr_mRNA_mixed, _ = pearsonr(mRNA, mixed)
    #print('mRNA/mixed Pearsons correlation: %.3f' % pear_corr_mRNA_mixed)
    pear_corr_siRNA_mixed, _ = pearsonr(siRNA, mixed)
    #print('siRNA/mixed Pearsons correlation: %.3f' % pear_corr_siRNA_mixed)

    spear_corr_mRNA_siRNA, _ = spearmanr(mRNA, siRNA)
    #print('mRNA/siRNA Spearman correlation: %.3f' % spear_corr_mRNA_siRNA)
    spear_corr_mRNA_mixed, _ = spearmanr(mRNA, mixed)
    #print('mRNA/mixed Spearman correlation: %.3f' % spear_corr_mRNA_mixed)
    spear_corr_siRNA_mixed, _ = spearmanr(siRNA, mixed)
    # print('siRNA/mixed Spearman correlation: %.3f' % spear_corr_siRNA_mixed)

    #Pearson Plot
    labels = ['mRNA/siRNA','mRNA/mixed','siRNA/mixed']
    pearson_arr = [pear_corr_mRNA_siRNA, pear_corr_mRNA_mixed, pear_corr_siRNA_mixed]
    spearman_arr = [spear_corr_mRNA_siRNA, spear_corr_mRNA_mixed, spear_corr_siRNA_mixed]
    axs[i].bar(X, pearson_arr, color = 'b', label = 'pearson coefficient', width = 0.25)
    axs[i].bar(X + 0.25, spearman_arr, color = 'g', label = 'spearman coefficient', width = 0.25)
    axs[i].title.set_text(feat_type + ' Features')
    
    #plt.title(feat_type)
    plt.xticks([r + 0.125 for r in range(3)], labels)
    axs[i].set_xticks([r + 0.125 for r in range(3)]) 
    axs[i].set_xticklabels(labels, fontsize=12)
    #axs[i].set_xticklabels([r + 0.125 for r in range(3)], labels)
    axs[i].set_ylim([-1, 1])
    axs[i].set_ylabel('Correlation Coefficients')
    axs[i].legend(loc="upper left", bbox_to_anchor=(1,1))
    
# for ax in axs.flat:
#     ax.label_outer()
#plt.ylabel('Correlation Coefficients')
plt.show()
```
