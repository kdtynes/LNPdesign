# Drug-Based Preprocessing

## File Description

In this jupyter notebook, we will preprocess the input LNP data and employ feature engineering to generate Mordred descriptors and interaction energies for each sample. Next, we will separate the data into different subsets based on payload: mRNA, siRNA, bcDNA and perform a test/train split on each. Lastly, we will combine each of the three training and testing datasets for the multi-drug model.

## Data preprocessing

```python
# Data preprocessing
def data_preprocessing(data):
    data.replace([np.inf, -np.inf], np.nan,inplace=True)
    data.fillna(0, inplace=True)

    # Renaming columns
    data = data.rename(columns = {'Included in screen? ': 'classifier'})
    data = data.rename(columns = {'Helper Lipid': 'HelperLipid'})
    data = data.rename(columns = {'Diameter (nm)': 'diameter'})

    #Filter Columns #'level_0', 
    data = data.drop(columns=['LNP #','Date','Initials','Project',
                      'Screen Successful?','Confirmed?','Confirmation Success','Cell Types'], errors='ignore')
    # data = data.drop(columns=['LNP #','diameter','Date','Initials','Project',
    #                   'Screen Successful?','Confirmed?','Confirmation Success','Cell Types'])
    # data = data.drop(columns =['Lipomer SMILES', 'Cholesterol SMILES', 
    #                            'Helper Lipid SMILES', 'PEG SMILES'])
    #data = data.drop(columns = ['Total','Elst','Exch','Ind','Disp']) #Test w/o interactions

    # Removing the C from the PEGChain 
    # data['PEGChain'] = data['PEGChain'].str.replace('C', '')
    data['PEGChain'] = data['PEGChain'].str.strip()
    data['Payload'] = data['Payload'].str.strip()
    data['classifier'] = data['classifier'].str.strip()

    # Converting to binary classification {0, 1}
    data.loc[data.classifier == 'Y', 'classifier'] = 1
    data.loc[data.classifier == 'N', 'classifier'] = 0

    # # Removing the points with "-" (no information)
    # data = data[data.HelperLipid != '-']
    data = data[data.classifier.isnull() == False]
    # # removing the NaNs
    # data = data[data.diameter.isnull()==False]

    # In the HelperLipid column '-' are equivalent to 'None' (Helper Lipid Mole % = 0)
    data.loc[data.HelperLipid == '-', 'HelperLipid'] = 'None'
    #Replace Null Helper Lipids with Lipomers
    data.loc[data.HelperLipid == 'None', 'HelperLipid'] = data.Lipomer
    data = data.reset_index(drop = True)

    #Use Ordinal Encoding on Payload
    data.loc[data['Payload'].str.contains('mRNA'), 'Payload'] = "0" #mRNA
    data.loc[data['Payload'].str.contains('aVHH'), 'Payload'] = "0"
    data.loc[data['Payload'].str.contains('si'), 'Payload'] = "1" #siRNA
    data.loc[data['Payload'].str.contains('BC'), 'Payload'] = "2" #DNA
    data['Payload'] = pd.to_numeric(data['Payload'], errors='ignore')

    #Separate data and labels
    y = np.array(data['classifier'])
    data = data.drop(['classifier'], 1)
    N = data.shape[0]
    D = data.shape[1]

    #Check Data Imbalance:
    vc = np.unique(y, return_counts = True)
    for i,num in enumerate(vc[0]):
        print('Number of '+str(vc[0][i])+'\'s: '+str(vc[1][i])+ ' (' + str(round(100*vc[1][i]/N, 1)) + '%)')

    print("Number of data points: ", N, " :: Number of features: ", D)
    print()
    print("Feature names : ", data.columns)
    return data, y
data, y = data_preprocessing(data)
```

```python
def add_smiles(data,smiles):
    PEGMW = np.unique(data.iloc[:,3])
    print(PEGMW)
    PEGChain = np.unique(data.iloc[:,4])
    print(PEGChain)
    cols_to_smiles = ['Lipomer','Cholesterol','HelperLipid', 'PEG']
    data['PEG'] = pd.Series()
    data['PEG'] = data['PEGChain'] + 'PEG' + data['PEG MW'].astype(str)
    for col in cols_to_smiles:
        data = data.join(smiles.set_index('Name'), on=col)
        print('---',col,'SMILES Unknown---')
        print(np.unique(data.loc[data.SMILES.isnull(),col]))
        new_name = col + ' SMILES'
        data = data.rename(columns = {'SMILES': new_name})

    print('---Number of rows without any NaN SMILES---')
    nonnull = data[data['Lipomer SMILES'].isnull() == False]
    nonnull = nonnull[nonnull['Cholesterol SMILES'].isnull() == False]
    nonnull = nonnull[nonnull['HelperLipid SMILES'].isnull() == False]
    nonnull = nonnull[nonnull['PEG SMILES'].isnull() == False]
    print(len(nonnull))
    data = nonnull.reset_index(drop = True)
    return data

data = add_smiles(data,Smiles)
#data = data.drop(columns = ['Lipomer', 'Cholesterol', 'HelperLipid', 'PEGChain', 'PEG MW', 'PEG'])
data

#Export Smiles
#data.iloc[:,-4:].to_csv("../../Datasets/data_smiles_complete.csv", index=False)
```

```python
#Join interaction energies to dataset
def add_interactions(data, interactions):
    joined_data = pd.merge(data, interactions,  how='left', left_on=['Lipomer SMILES','Cholesterol SMILES'], right_on = ['A','B'])
    joined_data = joined_data.drop(columns = ['A','B','RA','ZA','RB','ZB'], errors='ignore')
    data = joined_data
    return data
data = add_interactions(data, interactions)
data
```

```python
#Export & Import Mordred Descriptors
#mordred_df.to_csv("../../Datasets/mordred_df.csv", index = False)
mordred_df = pd.read_csv('../../Datasets/mordred_df.csv')
mordred_df
```

```python
#Find mordred descriptors of every molecule first, and then append to dataset to prevent repeat work
def add_mordred(data, mordred_df):
    #Join dataset with mordred data
    cols_to_smiles = ['Lipomer','Cholesterol','HelperLipid', 'PEG']
    for i,col in enumerate(cols_to_smiles):
        feat = col + ' SMILES'
        print(feat)

        #filter errors in parsing, remove those samples
        #filter = [k is not None for k in mols]
        #mols = [i for i in mols if i is not None]
        #data = data.loc[np.array(filter)]
        temp = mordred_df.copy()
        temp = temp.drop(columns = ['Name'])
        temp = temp.drop_duplicates(subset="SMILES")

        #rename columns & join
        drop = temp['SMILES']
        temp = temp.drop(columns = ['SMILES'])
        temp = temp.add_suffix('_' + col) 
        temp.insert(0, 'SMILES', drop)
        data = data.join(temp.set_index('SMILES'), how='left', on=feat)
    data = data.drop(columns =['Lipomer SMILES', 'Cholesterol SMILES', 
                                  'HelperLipid SMILES', 'PEG SMILES'])
    return data

data = add_mordred(data, mordred_df)
```

```python
#Remove columns with Mordred Errors
#convert object type into bool
def remove_mordred_errors(data):
    #print('data types',np.array(data.dtypes))
    types, type_idx = np.unique(data.dtypes,return_inverse = True)
    #print('types',types)
    obj_idx = np.argwhere(type_idx == np.where(types == np.object)[0][0])
    #print('obj indices',obj_idx)
    obj_idx = [i[0] for i in obj_idx if i[0] > 11]
    #print('obj indices',obj_idx)
    cols_to_remove = []
    for i in obj_idx:
        data[data.columns[i]] = pd.to_numeric(data[data.columns[i]], errors='coerce').fillna(0)
        if (data[data.columns[i]] == 0).all():
            cols_to_remove.append(data.columns[i])
    #print('cols to remove',cols_to_remove)
    print('num cols removed',len(cols_to_remove))
    data = data.drop(cols_to_remove, axis = 1)
    data.replace([np.inf, -np.inf], np.nan, inplace=True)
    return data.fillna(0)
data= remove_mordred_errors(data)
data
```

## Split Into Drug-Based Datasets

```python
def train_test_split(data):
    N = data.shape[0]
    train_N = int(np.ceil(0.8*N))
    
    random.seed(33333)
    np.random.seed(33333)
    shuffle_data = np.arange(N)
    random.shuffle(shuffle_data)

    training_rows = shuffle_data[:train_N]
    testing_rows = shuffle_data[train_N:]

    data_train = data.iloc[training_rows].reset_index(drop = True)
    data_test = data.iloc[testing_rows].reset_index(drop = True)
    return data_train, data_test
```

```python
def check_imbalance(data):
    vc = np.unique(data['label'], return_counts = True)
    N = data.shape[0]
    D = data.shape[1]

    for i,num in enumerate(vc[0]):
        print('Number of '+str(vc[0][i])+'\'s: '+str(vc[1][i])+ ' (' + str(round(100*vc[1][i]/N, 1)) + '%)')

    print("Number of data points: ", N, " :: Number of features: ", D)
```

```python
mRNA_data = data.loc[data['Payload'] == 0] #mRNA
siRNA_data = data.loc[data['Payload'] == 1] #siRNA
bcDNA_data = data.loc[data['Payload'] == 2] #DNA
```

#### mRNA

```python
check_imbalance(mRNA_data)
mRNA_train, mRNA_test = train_test_split(mRNA_data)

#Export
mRNA_train.to_csv("../../Datasets/Drug-Based/mRNA_train.csv", index=False)
mRNA_test.to_csv("../../Datasets/Drug-Based/mRNA_test.csv", index=False)

mRNA_data
```

#### siRNA

```python
check_imbalance(siRNA_data)
siRNA_train, siRNA_test = train_test_split(siRNA_data)

#Export
siRNA_train.to_csv("../../Datasets/Drug-Based/siRNA_train.csv", index=False)
siRNA_test.to_csv("../../Datasets/Drug-Based/siRNA_test.csv", index=False)

siRNA_data
```

#### bcDNA

```python
check_imbalance(bcDNA_data)

bcDNA_train, bcDNA_test = train_test_split(bcDNA_data)

#Export
bcDNA_train.to_csv("../../Datasets/Drug-Based/bcDNA_train.csv", index=False)
bcDNA_test.to_csv("../../Datasets/Drug-Based/bcDNA_test.csv", index=False)

bcDNA_data
```

#### Multidrug

```python
#Concat all three datasets
multidrug_train = pd.concat([mRNA_train, siRNA_train, bcDNA_train], ignore_index=True)
multidrug_test = pd.concat([mRNA_test, siRNA_test, bcDNA_test], ignore_index=True)

#Shuffle
multidrug_train = multidrug_train.sample(frac=1, random_state = 33333).reset_index(drop=True)
multidrug_test = multidrug_test.sample(frac=1, random_state = 33333).reset_index(drop=True)

check_imbalance(pd.concat([multidrug_train, multidrug_test], ignore_index=True))

#Export
multidrug_train.to_csv("../../Datasets/Drug-Based/multidrug_train.csv", index=False)
multidrug_test.to_csv("../../Datasets/Drug-Based/multidrug_test.csv", index=False)

multidrug_train
```
